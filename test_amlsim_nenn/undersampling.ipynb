# Realizando o Undersampling da Base de Dados

Fazendo um undersampling simples para equilibrar a presen√ßa de fraudes na base de dados em 50%.

```python
import pandas as pd
```

```python
transactions = pd.read_csv('new_1M/transactions.csv')
```

```python
transactions
```

```python
len(transactions[transactions['isSAR'] == True])
```

```python
sample_falses = transactions[transactions['isSAR'] == False].sample(1039)
```

```python
len(sample_falses.sort_values(by=['step']).step.unique())
```

```python
transactions[(transactions.index.isin(sample_falses.index))]
```

```python
sample_falses = sample_falses.reset_index()
transactions = transactions.reset_index()
```

```python
balanced = transactions[(transactions['isSAR'] == True) | (transactions.index.isin(sample_falses.index))]
```

```python
balanced
```

```python
balanced_account_set = set(balanced['nameOrig'].values.tolist() + balanced['nameDest'].values.tolist()) 
```

```python
len(balanced_account_set)
```

```python
accounts = pd.read_csv('new_1M/accounts.csv')
```

```python
accounts
```

```python
new_account_set = accounts[accounts['acct_id'].isin(balanced_account_set)]
```

```python
new_account_set
```

```python
new_account_set = new_account_set.reset_index(drop=True).reset_index()
```

```python
new_account_set = new_account_set.rename(columns={'index': 'NEW_ACC_ID'})
```

```python
new_account_set['NEW_CUSTOMER_ID'] = 'C_' + new_account_set['dsply_nm'].astype(str)
```

```python
new_account_set
```

```python
keys = new_account_set.acct_id.values
values = new_account_set.NEW_ACC_ID.values

parse_keys = {}

for k, v in zip(keys, values):
    parse_keys[k] = v
```

```python
def convert_to_new(line):
    line['nameOrig'] = parse_keys[line['nameOrig']]
    line['nameDest'] = parse_keys[line['nameDest']]
    return line
```

```python
balanced = balanced.apply(convert_to_new, axis=1)
```

```python
balanced
```

```python
new_account_set.acct_id = new_account_set.NEW_ACC_ID
new_account_set.dsply_nm = new_account_set.NEW_CUSTOMER_ID
```

```python
new_account_set = new_account_set.drop(['NEW_ACC_ID', 'NEW_CUSTOMER_ID'],axis=1)
```

```python
new_account_set.to_csv('new_1M/accounts_undersampling.csv')
balanced.to_csv('new_1M/transactions_undersampling.csv')
```
